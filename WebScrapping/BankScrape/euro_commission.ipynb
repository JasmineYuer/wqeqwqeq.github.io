```python
#再爬所有
for i in final_links:
        uClient1 = uReq(i)
        page_html1 = uClient1.read()
        uClient1.close()
        page_soup1 = soup(page_html1, "html.parser")
        file=docx.Document()
        name=page_soup1.find_all('title')[0].string
        file.add_heading(page_soup1.find_all('title')[0].string,0)
        llst=[]

        for i in page_soup1.h2.parent.find_all('p'):
            
            if i.string==None:
                try:
                    k=str(i).split('</span>\n')[1].split('\n<span')[0]
                    llst.append(k)
                except:
                    pass
                llst.append(page_soup1.find_all('footer')[0].string)
            else:
                llst.append(i.string)



        a_lst=[]
        for i in page_soup1.h2.next_siblings: 
            a_lst.append(i.string)
        final_list=[]
        for i in llst:
            if i not in a_lst:
                final_list.append(i)

        for i in final_list:
            file.add_paragraph(i)
        file.add_paragraph(page_soup1.find('h2').string)
        for i in a_lst:
            file.add_paragraph(i)
        try:
            file.save(path+'/'+name+'.docx')
        except:
            file.save(path+'/'+name[0:3]+'.docx')
        print('finished {}'.format(name))
```
