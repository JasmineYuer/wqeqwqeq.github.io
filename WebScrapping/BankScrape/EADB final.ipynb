```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
import pandas as pd

my_url = 'http://eadb.org/project/'
hah='/page/'
url_list = [my_url + hah + str(a) + '/' for a in range(1,5)]
```

```python
big_table = pd.DataFrame()
for i in url_list: 
    req = Request(i, headers={'User-Agent': 'Mozilla/5.0'})
    page_html = urlopen(req).read()
    page_soup = soup(page_html, "html.parser")
    links = [a['href'] for a in page_soup.find_all('a',href = True)]
    for link in set(links):
        if 'eadb.org/project/' in link:
            req = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
            page_html = urlopen(req).read()
            page_soup = soup(page_html, "html.parser")
            #Find Des
            qee = page_soup.find_all('p')[:-1]
            Description = ''
            for i in qee:
                try:
                    if i == None:
                        continue
                    else:
                        Description = Description + i.string
                except:
                    pass
            #Find Client
            try:
                if page_soup.find_all('div',{'class':['portfolio-term']})[0].string == None:
                    continue
                else:
                    Client = page_soup.find_all('div',{'class':['portfolio-term']})[0].string
            except:
                pass
            #Find Date
            try:
                if page_soup.find_all('div',{'class':['portfolio-term']})[1].string == None:
                    continue
                else:
                    Date = page_soup.find_all('div',{'class':['portfolio-term']})[1].string
            except:
                pass
            #Find Location
            try:
                if page_soup.find_all('div',{'class':['portfolio-term']})[2].string == None:
                    continue
                else:
                    Location = page_soup.find_all('div',{'class':['portfolio-term']})[2].string
            except:
                pass
            
            #Find Category
            try:
                if page_soup.find_all('div',{'class':['portfolio-term']})[3].string == None:
                    continue
                else:
                    Category = page_soup.find_all('div',{'class':['portfolio-term']})[3].string
            except:
                pass
            #Create Table
            table=pd.DataFrame([[Description,Client,Date,Location,Category]],columns=['Description','Client','Date','Location','Category'])
            big_table=pd.concat([big_table,table],axis=0)
```

```python
big_table.index=range(len(big_table))
big_table
```
